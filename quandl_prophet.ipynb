```python
fb = fb.reset_index()
fb = fb.rename(columns={'Date': 'ds', 'Adj. Close': 'y'})
fb
```

```python
fb = fb.ix[:, ['ds', 'y']]
fb.head()
```

```python
import fbprophet
# Make the prophet models and fit on the data
# changepoint_prior_scale can be changed to achieve a better fit
fb_prophet = fbprophet.Prophet(changepoint_prior_scale=0.05)
fb_prophet.fit(fb)
```

```python
# Make a future dataframe for next years
fb_forecast = fb_prophet.make_future_dataframe(periods=365*1, freq='D')
# Make predictions
fb_forecast = fb_prophet.predict(fb_forecast)
```

```python
fb_prophet.plot(fb_forecast, xlabel = 'Year', ylabel = 'Stock Price($)')
plt.title('Facebook Stock Projection')
```
